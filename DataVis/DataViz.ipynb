```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../Data/heart_2020_cleaned.csv")
```

```python
df.head()
```

# Dataset analysis :

```python
obj1 = df.columns
i = 1
plt.figure(figsize = (25,25))
for col in obj1:
    plt.subplot(6,3,i)
    sns.histplot(df, x=col)
    #countplot(x = col ,hue = df[col] , data = df)
    i +=1
```

# Data distribution of the target y = 'HeartDisease'

```python
plt.figure(figsize=(7,7))
sns.countplot(data=df,x="HeartDisease")
```

```python
obj1 = df.select_dtypes(include = object).columns
i = 1
plt.figure(figsize = (15,25))
for col in obj1:
    plt.subplot(6,3,i)
    sns.countplot(x = 'Sex',hue = df[col] , data = df)
    i +=1
```

# Dataset outliers :

```python
float1 = df.select_dtypes(exclude = object).columns
i=1
plt.figure(figsize=(15,15))
for col in float1:
    plt.subplot(3,2,i)
    sns.boxplot(y=df[col], x = df['Race'])
    i+=1
```

# Data distribution per general health

```python
i=1
plt.figure(figsize=(15,15))
for col in float1:
    plt.subplot(3,2,i)
    sns.kdeplot(x = df[col], hue = df['GenHealth'])
    i+=1
```

```python
g = sns.FacetGrid(df, col='HeartDisease', hue="GenHealth", col_wrap=4)
g.map(sns.kdeplot, 'SleepTime',alpha=.7)
plt.figure(figsize=(15,7))
g.add_legend()
```

# Distribution on race :

```python
num_race = df['Race'].value_counts()
num_race.plot(kind='pie', autopct='%.2f', figsize=(6, 6))
```

```python
plt.figure(figsize=(20,10))
sns.histplot(df, x="Race")
```

# Distribution on sex :

```python
num_male_female = df['Sex'].value_counts()
num_male_female.plot(kind='pie', autopct='%.2f', figsize=(6, 6))
```

# Distribution on age category :

```python
plt.figure(figsize=(15,7))
sns.countplot(data=df,x="AgeCategory")
```
