```python
# to find defferences classes of col
from sklearn import preprocessing
import numpy as np
le = preprocessing.LabelEncoder()
y = df["GenHealth"]
le.fit(y)
le.classes_
```

### Convert object features to int

```python
array = ["HeartDisease","Smoking","AlcoholDrinking","Stroke","DiffWalking","PhysicalActivity","Asthma","KidneyDisease","SkinCancer"]
df[array] = df[array].replace({"No": 0, "Yes": 1})
df["Race"] = df["Race"].replace({
    "American Indian/Alaskan Native": 0, 
    "Asian": 1,
    "Black": 2,
    "Hispanic" : 3,
    "Other" : 4,
    "White" : 5
})
df["Diabetic"] = df["Diabetic"].replace({
    "No": 0, 
    "No, borderline diabetes": 1,
    "Yes": 2,
    "Yes (during pregnancy)" : 3
})
df["Sex"] = df["Sex"].replace({
    "Female": 0, 
    "Male": 1
})
df["AgeCategory"]=df["AgeCategory"].replace({
    "18-24" : 1,
    "25-29" : 2,
    "30-34" : 3,
    "35-39" : 4,
    "40-44" : 5,
    "45-49" : 6,
    "50-54" : 7,
    "55-59" : 8,
    "60-64" : 9,
    "65-69" : 10,
    "70-74" : 11,
    "75-79" : 12,
    "80 or older" : 13,
})
df["GenHealth"] = df["GenHealth"].replace({
    "Poor": 1,
    "Fair": 2,
    "Good" : 3,
    "Very good" : 4,
    "Excellent" : 5
})
```
