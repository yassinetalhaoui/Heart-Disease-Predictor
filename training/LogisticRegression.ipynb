```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import pickle
```

```python
df = pd.read_csv("../Data/heart_2020_cleaned_for_models.csv")
```

```python
df.head()
```

```python
y = df["HeartDisease"]
X = df.drop("HeartDisease", axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = LogisticRegression(max_iter = 500)
model.fit(X_train, y_train)  # apprentissage
y_test_pred = model.predict(X_test) # predictionde test
```

```python
test_score = model.score(X_test, y_test)
print("Le score en test est {}".format(test_score))
```

```python
print(metrics.classification_report(y_test, y_test_pred))
```

```python
from sklearn.model_selection import GridSearchCV
model=LogisticRegression(max_iter= 500)
grid={'penalty': ["l1","l2","elasticnet","none"],'max_iter': [500,750,1000]}
model_cv=GridSearchCV(model,grid,cv=3)
model_cv.fit(X_train,y_train)
print("tuned hpyerparameters :(best parameters) ",model_cv.best_params_)
print("accuracy :",model_cv.best_score_)
```

```python
model = LogisticRegression(max_iter = 500,penalty = 'l2')
model.fit(X_train, y_train)  # apprentissage
y_test_pred = model.predict(X_test) # predictionde test
```

```python
test_score = model.score(X_test, y_test)
print("Le score en test est {}".format(test_score))
```

```python
filename = '../models/LogisticRegression.pkl'
pickle.dump(model, open(filename, 'wb'))
```
