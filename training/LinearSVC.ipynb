```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
import pickle
```

```python
df = pd.read_csv("../Data/heart_2020_cleaned_for_models.csv")
df.head()
```

```python
y = df["HeartDisease"]
X = df.drop("HeartDisease", axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
C = 10.0 # paramètre de régularisation
model = svm.LinearSVC(C=C)
model.fit(X_train, y_train)
y_test_pred = model.predict(X_test) # predictionde test
```

```python
test_score = model.score(X_test, y_test)
print("Le score en test est {}".format(test_score))
```

```python
print(metrics.classification_report(y_test, y_test_pred))
```

```python
from sklearn.model_selection import GridSearchCV
model=svm.LinearSVC()
grid={'penalty': ["l1","l2"],'C': [1,3,5,7,8,10,15,20]}
model_cv=GridSearchCV(model,grid,cv=3)
model_cv.fit(X_train,y_train)
print("tuned hpyerparameters :(best parameters) ",model_cv.best_params_)
print("accuracy :",model_cv.best_score_)
```

```python
C = 7.0 # paramètre de régularisation
model = svm.LinearSVC(C=C)
model.fit(X_train, y_train)
y_test_pred = model.predict(X_test) # predictionde test
```

```python
test_score = model.score(X_test, y_test)
print("Le score en test est {}".format(test_score))
```
